```python
# Bibliotecas
import numpy as np
import matplotlib.pyplot as plt
from scipy import stats, special
import copy
```

## 1) CDF da Distribuição Normal

* Função: `my_normal_distribution_cdf`
* Caso deseje p(x1..x2), construa uma lista como argumento de entrada \[x1, x2\]
* Testada com $\mu$, $\sigma$ = 0.0, 1.0 e $\mu$, $\sigma$ = 2.0, 2.0
* Comparada com: CDF do scipy (`scipy.stats.norm.cdf`)

```python
def my_normal_distribution_cdf(x, mu = 0, std_dev = 1):
    # Caso x seja uma lista, use-o normalmente
    # Caso x não seja uma lista, transforme-o em [-inf, x]
    if type(x) != type([]):
        x = [float("-inf"), x]
    
    p1 = 0.5 + special.erf( (x[0]-mu)/(std_dev*np.sqrt(2)) )/2
    p2 = 0.5 + special.erf( (x[1]-mu)/(std_dev*np.sqrt(2)) )/2
    
    return p2-p1
```

```python
x = np.linspace(-10, 10, 1000)

y1 = my_normal_distribution_cdf(x)
y2 = my_normal_distribution_cdf(x, mu = 2, std_dev = 2)
```

```python
plt.plot([0, 0], [0, 1], 'k--')
plt.plot([2, 2], [0, 1], 'k--')
plt.plot([-10, 10], [0, 0], 'k--')
plt.plot(x, y1, label="$\mu$ = 0 ; $\sigma$ = 1")
plt.plot(x, y2, label = "$\mu$ = 2 ; $\sigma$ = 2")
plt.legend()
plt.show()
```

```python
# Comparação com scipy
ys = stats.norm.cdf(x)
```

```python
plt.plot([0, 0], [0, 1], 'k--')
plt.plot([-10, 10], [0, 0], 'k--')
plt.plot(x, y1, label='my_normal_dist_cdf')
plt.plot(x, ys, label='scipy.stats.norm.cdf')
plt.legend(loc=2)
plt.show()
```

```python
# Testando para intervalos
print('Para \u03BC = 0 e \u03C3 = 1 : p([-1, 1]) = %.5f ' % my_normal_distribution_cdf([-1, 1]))
print('Para \u03BC = 2 e \u03C3 = 2 : p([ 2, 6]) = %.5f ' % my_normal_distribution_cdf([2, 6], 2, 2))
```
