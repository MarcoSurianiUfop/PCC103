```python
# Bibliotecas
import numpy as np
import matplotlib.pyplot as plt
from scipy import stats
```

## 1) PDF da Distribuição Normal

* Função: `my_normal_distribution_pdf`
* Testada com $\mu$, $\sigma$ = 0.0, 1.0 e $\mu$, $\sigma$ = 2.0, 2.0
* Comparada com: PDF do scipy (`scipy.stats.norm.pdf`)

```python
def my_normal_distribution_pdf(x, mu = 0, std_dev = 1):
    return np.exp( -1*(x - mu)**2/(2*std_dev**2) ) / (std_dev*np.sqrt(2*np.pi))
```

```python
x = np.linspace(-10, 10, 1000)

y1 = my_normal_distribution_pdf(x)
y2 = my_normal_distribution_pdf(x, mu = 2, std_dev = 2)
```

```python
plt.plot([0, 0], [0, 0.45], 'k--')
plt.plot([2, 2], [0, 0.45], 'k--')
plt.plot([-10, 10], [0, 0], 'k--')
plt.plot(x, y1, label="$\mu$ = 0 ; $\sigma$ = 1")
plt.plot(x, y2, label = "$\mu$ = 2 ; $\sigma$ = 2")
plt.legend()
plt.show()
```

```python
# Comparação com scipy
ys = stats.norm.pdf(x)
```

```python
plt.plot([0, 0], [0, 0.45], 'k--')
plt.plot([-10, 10], [0, 0], 'k--')
plt.plot(x, y1, label='my_normal_dist_pdf')
plt.plot(x, ys, label='scipy.stats.norm.pdf')
plt.legend(loc=2)
plt.show()
```
